```python
hamlet = nltk.corpus.gutenberg.raw('shakespeare-hamlet.txt')
print(len(hamlet))
```

```python
tokenized_text = word_tokenize(hamlet)
text_ner3 = st3.tag(tokenized_text)
text_ner4 = st4.tag(tokenized_text)
text_ner7 = st7.tag(tokenized_text)

print(text_ner3)
print(text_ner4)
print(text_ner7)
```

```python
from itertools import groupby

class3 = []
class4 = []
class7 = []

print("**** 3 classes ****")
for tag, chunk in groupby(text_ner3, lambda x:x[1]):
    if tag == "PERSON":
        for w, t in chunk:
            if w not in class3:
                class3.append(w)

print(class3)
        
print("**** 4 classes ****")
for tag, chunk in groupby(text_ner4, lambda x:x[1]):
    if tag == "PERSON":
        for w, t in chunk:
             if w not in class4:
                class4.append(w)
        
print(class4)        
        
print("**** 7 classes ****")
for tag, chunk in groupby(text_ner7, lambda x:x[1]):
    if tag == "PERSON":
        for w, t in chunk:
            if w not in class7:
                class7.append(w)
            
print(class7)
```

```python
from nltk.corpus import stopwords, reuters
docNames = reuters.fileids()
test = [x for x in docNames if x.startswith('test/')]
train = [x for x in docNames if x.startswith('training/')]

#train_data = [reuters.raw(category) for category in train]
#test_data = [reuters.raw(category) for category in test] 
```

```python
train_data = [reuters.raw(category) for category in train]
test_data = [reuters.raw(category) for category in test] 
```

```python
train_data
```

```python
print(len(train_data))
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

filterWords = stopwords.words("english")

def tokenize(text):
    min_length = 3
    words = [word.lower() for word in word_tokenize(text) if word not in filterWords]
    tokens = [PorterStemmer().stem(token) for token in words]
    p = re.compile('[a-zA-Z0-9]+');
    tokens = [token for token in tokens if p.match(token) and len(token) >= min_length]
    return tokens;
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words=filterWords,
                             tokenizer=tokenize)
```

```python
vectorised_train_documents = vectorizer.fit_transform(train_data)
vectorised_test_documents = vectorizer.transform(test_data)
```

```python
vectorised_train_documents
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
train_labels = mlb.fit_transform([nltk.corpus.reuters.categories(docName)
                                  for docName in train])
test_labels = mlb.transform([nltk.corpus.reuters.categories(docName)
                             for docName in test])
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
classifier = OneVsRestClassifier(LinearSVC(random_state=42))
classifier.fit(vectorised_train_documents, train_labels)
 
predictions = classifier.predict(vectorised_test_documents)
```

```python
predictions
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
precision = precision_score(test_labels, predictions,
                            average='micro')
recall = recall_score(test_labels, predictions,
                      average='micro')
f1 = f1_score(test_labels, predictions, average='micro')
 
print("Micro-average quality numbers")
print("Precision: {:.4f}, Recall: {:.4f}, F1-measure: {:.4f}"
        .format(precision, recall, f1))
 
precision = precision_score(test_labels, predictions,
                            average='macro')
recall = recall_score(test_labels, predictions,
                      average='macro')
f1 = f1_score(test_labels, predictions, average='macro')
 
print("Macro-average quality numbers")
print("Precision: {:.4f}, Recall: {:.4f}, F1-measure: {:.4f}".format(precision, recall, f1))
```
